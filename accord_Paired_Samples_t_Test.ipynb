## Part II. Background

<details>
<summary>  ðŸ“š  Click for recommended background reading </summary>

A paired samples *t*-test, also known as the dependent samples t-test, compares the means of two related groups under different conditions, using hypothesis testing to determine if there is a significant difference between the measurements.

This test is applicable when the same participants are measured twice under different conditions or when there are pairs of matched subjects. Put another way, we can use a paired samples *t*-test in one of the following situations:  

- We have two sets of related data of a continuous, quantitative variable.
- The data comes from the same participants measured under two different conditions. For example, repeated measurements at two different points in time (e.g., before and after an intervention) where each participant will have two measures, time1 and time2.
- The data are from different participants that are carefully matched based on certain characteristics (e.g., key demographic variables) to ensuring that each participant in one condition has a corresponding counterpart in the other condition, allowing for comparison.

### 3. Decide on a One-Tailed or Two-Tailed Test

Now that you have seen a preview, it is time to use the data.

There are two ways to write hypotheses for *t*-tests. In a **one-tailed test**, we are testing for the presence of a **directional relationship**. For example, if we hypothesize that the mean life expectancy in 2018 will be greater than the mean life expectancy in 2008, we are predicting one direction. Similarly, if we hypothesized that mean life expectancy would be greater in 2008 compared to 2018, we would be prediciting an outcome in the other direction. The advantage is that our statistical test will be more sensitive (i.e., more statistical power) when running a one-tailed test. The disadvantage is that would not be able to test for *both* of these relationships at the same time.  

We would want to test for both if we do not strong have evidence that one measure is either more or less than the other. To look for a relationship without a direction, we need to use a **two-tailed test**. Two-tailed tests have the advantage of looking in both directions at a cost of some sensitivity (i.e., less statistical power).  


In this case, global trends have suggested that life expectancy across countries has been increasing across time. Medical, social, and economic advancements have been identified as factors that positively impact life expectancy. Therefore, we will use a one-tailed test in this example.


To summarize, use a one-tailed test when you have a reason to predict the direction of the relationship before you start. Otherwise, use a two-tailed test. For our data, we will use a one-tailed test to test whether mean life expectancy in 2018 is greater than 2008.


### 4. Write Hypotheses

In a paired t-test, we are testing the difference between related observations. The hypotheses are stated in terms of the differences between the pairs. For each pair, we calculate $D=xâˆ’x_1$â€‹, where:

* $D$ is the difference between the paired observations.
* $x$ is the value of the first observation (e.g., mean life expectancy in 2018).
* $x_1$ is the value of the second observation (e.g., mean life expectancy in 2008).

The mean of these differences, denoted as ${\mu_D}$, represents the average difference between the paired observations. The paired t-test then tests whether this mean difference is significantly different from zero.   

The one-tailed hypotheses for a paired samples t-test are:

> $H_0: {\mu_D} â€‹â‰¤ 0$

> $H_a: {\mu_D} â€‹> 0$


The null hypothesis ($H_0$) is that the mean difference in life expectancy between 2018 and 2008 is less than or equal to zero.

The alternative hypothesis ($H_a$, sometimes labeled $H_1$) is that the mean difference in life expectancy between 2018 and 2008 is greater than zero. The alternative hypothesis is always the **logical complement** to the null hypothesis. If the null is true, then the alternative is false. If the null is false, then the alternative is true.   

If we had predicted an outcome in the other direction (that mean life expectancy would be greater in 2008 compared to 2018), then we would simply switch the direction of the signs in the null and alternative hypotheses.   


Although we do not need these right now, here is how we would have written hypotheses for two-tailed tests. Notice how the null hypothesis is always the logical complement of the alternative hypothesis. In the context of comparing life expectancy in 2008 and 2018, a two-tailed test would examine whether the mean life expectancy in 2018 is significantly different from that in 2008, without specifying whether it is higher or lower.

> $H_0: {\mu_D} =  0$

> $H_a: {\mu_D} \ne 0$

### <font color="red">âž¡</font> Run the Cell to Conduct a Paired Samples *t*-Test  

```python
# Import library
from scipy import stats

# Paired samples t-test
results = stats.ttest_rel(
    data['exp2018'], data['exp2008'],
    alternative='greater'
    )

# Print results
print(
    f"""Results:
    t-statistic = {results.statistic}
    p-value = {results.pvalue}
    degrees of freedom = {results.df}
    """
    )
```

###6. Interpret the Results

Refer to the following to interpret the output above:

* The *t* statistic indicates the magnitude of the mean difference between the paired samples relative to its standard error.
* The *p*-value indicates whether the difference is statistically significant. Specifically, it represents the probability of observing this value of *t* if the null hypothesis were true (i.e., if there were no mean difference).
* The degrees of freedom represents the number of values in a calculation that are free to vary. For the paired samples *t*-test, this is $N - 1$.

### 7. Decide

Our *p*-value leads us to our decision.   

<details>
<summary>  ðŸ“š  Click for recommended background reading </summary>

When the mean difference is large, *t* is large and *p* is small (less than .05). This would mean a large enough mean difference was observed to get beyond the threshold of evidence. When this happens, we can claim there is a difference. The term for a mean difference that is large enough to reject the null hypothesis is a **statistically significant difference**.

When we retain the null, our results are inconclusive. Retaining the null gives us no evidence, which can be confusing. When the mean difference is small, *t* is small and *p* is large (greater than .05). If we observed a small mean difference that was not enough to get us over our threshold of evidence, we cannot claim that there is no difference at all. At the same time, we cannot claim that there is even a small difference. The point of Null Hypothesis Testing (NHST) is to give us a guideline for how big of a difference is needed before we can call it a reliable difference.   

In summary, only rejecting the null allows us to make a conclusion.

Only the *p*-value is needed to reject or retain. However, the values of *t*, *df*, mean difference, and Cohen's *d* (a measure of effect size) are useful values to add context to the *p*-value. For this reason, whenever you report a *p*-value to others, you should also include the values of *t*, *df*, mean difference, and Cohen's *d*.   


### 8. Mean Difference and Effect Size

<details>
<summary>  ðŸ“š  Click for recommended background reading </summary>

To quantify the average change between paired samples and assess the standardized effect size of the observed difference, we will calculate the mean difference and Cohen's *d* effect size. The mean difference is the average difference between our paired observations. Cohen's *d* provides a standardized measure of how large that difference is in standard deviations. These help us understand the practical significance and impact of the observed changes.   


The code in the next block will calculate these for you. For reference, these are the formulas for hand calculations.   

Mean Difference ($M_D$):

>$D = x - x_1$

>$M_D = \frac{\sum D}{N}$

* $x$ and $x_1$ are the paired observations
* $D$ is the difference for each pair
* $N$ is the number of pairs

Cohen's *d* ($d$):

>$d = \frac{M_D}{s_D}$


* $M_D$ is the mean difference
* $s_D$ is the standard deviation of the differences


### <font color="red">âž¡</font> Run the Cell to Calculate the Mean Difference and Cohen's *d*



```python
# Import library
import numpy as np

# Calulate mean difference
difference = data['exp2018'] - data['exp2008']
mean_diff = np.mean(difference)


# Calculate standard deviation of differences
std_diff = np.std(difference, ddof=1)

# Calculate Cohen's d
cohens_d = mean_diff / std_diff

# Print results
print(
    f"""Results:
    mean difference = {mean_diff}
    Cohen's d = {cohens_d}
    """
    )
```
